```python
import pandas as pd
import requests
import xmltodict
import time

# Load company list (symbol + issuer)
df = pd.read_csv("ind_nifty50list.csv")  # Must contain columns: symbol, issuer

# Set headers for requests
HEADERS = {
    "User-Agent": "Mozilla/5.0",
    "Accept": "*/*",
    "Accept-Language": "en-US,en;q=0.9",
    "Referer": "https://www.nseindia.com/companies-listing/corporate-filings-shareholding-pattern",
    "Connection": "keep-alive",
    "X-Requested-With": "XMLHttpRequest"
}

# Prepare session
session = requests.Session()
session.get("https://www.nseindia.com/companies-listing/corporate-filings-shareholding-pattern", headers=HEADERS)
time.sleep(1)

# To collect final results
records = []
dct = {}
# You need to set the curr quarter as the xml is different for the current quarter
curr_quarter = "30-JUN-2025"
for idx, row in df.iterrows():
    symbol = row['Symbol']
    issuer = row['Company Name']
    print(f"üöÄ Processing {symbol}...")

    try:
        api_url = f"https://www.nseindia.com/api/corporate-share-holdings-master?index=equities&symbol={symbol}&issuer={issuer.replace(' ', '%20')}"
        r = session.get(api_url, headers=HEADERS)
        r.raise_for_status()
        entries = r.json()

        # Select latest XBRL per quarter
        quarter_data = {}
        for entry in entries:
            period = entry['date']
            actions = entry.get('actions', [])
            if entry['xbrl'].endswith(".xml") :
                quarter_data[period] = entry['xbrl']


        for period, xbrl_url in quarter_data.items():
            # time.sleep(5)
            xbrl_resp = session.get(xbrl_url, headers=HEADERS)
            xbrl_resp.raise_for_status()
            xbrl_dict = xmltodict.parse(xbrl_resp.content)
            total = 0
            shares = xbrl_dict['xbrli:xbrl']['in-bse-shp:NumberOfFullyPaidUpEquityShares']
            for share in shares:
                # print(type(share))
                if share["@contextRef"] == 'PublicShareholdingI' or (period == curr_quarter and share["@contextRef"] == "PublicShareholding_ContextI"):
                    total = int(share["#text"].replace(",", ""))

            records.append({
                "symbol": symbol,
                "issuer": issuer,
                "quarter": period,
                "outstanding_shares": total
            })

    except Exception as e:
        print(f"‚ùå Failed for {symbol}: {e}")
    # time.sleep(1)

# Save result
final_df = pd.DataFrame(records)
final_df.to_csv("nifty50_outstanding_shares_quarterly.csv", index=False)
print("‚úÖ Done. Saved to nifty50_outstanding_shares_quarterly.csv")
```
